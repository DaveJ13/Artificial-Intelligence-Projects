```python
import numpy as np
import pandas as pd
import tflearn
from tensorflow.keras.utils import to_categorical
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer(as_frame = True)
print(data.DESCR)
```

```python
data.frame
```

```python
X = data.frame.iloc[:,0:20]
Y = data.frame["target"]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
x = X.apply(lambda x: ((x - x.min()) / (x.max() - x.min())))
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test, x_train, y_test, y_train = train_test_split(x, Y,  test_size = .33)
```

```python
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
x_train = x_train.astype("float32")
x_test = x_test.astype("float32")
```

```python
output = 2
y_train = to_categorical(y_train, output)
y_test = to_categorical(y_test, output)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
net = tflearn.input_data(shape = [None, 20])
net = tflearn.fully_connected(net, 32)  
net = tflearn.fully_connected(net, 32)
net = tflearn.fully_connected(net, 2, activation = "softmax") 
net = tflearn.regression(net) 
```

```python
model = tflearn.DNN(net)
model.fit(x_train, y_train, n_epoch = 10, batch_size = 16, show_metric = True)
```

```python
pred = model.predict(x_test)
```
