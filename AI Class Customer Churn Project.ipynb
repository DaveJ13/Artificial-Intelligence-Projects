```python
from keras import models
from keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from keras.datasets import mnist
from keras.utils.vis_utils import model_to_dot
from IPython.display import SVG
```

```python
churn
```

```python
X = pd.DataFrame(churn["tenure"])
Y = pd.DataFrame(churn["Churn"])
```

```python
Y.replace(("Yes", "No"), (1,0), inplace = True)
```

```python
print(X.shape)
print(Y.shape)
```

```python
x = X.apply(lambda x: ((x - x.min()) / (x.max() - x.min())))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test, x_train, y_test, y_train = train_test_split(X, Y,  test_size = .2)
```

```python
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
x_train = x_train.astype("float32")
x_test = x_test.astype("float32")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape
```

```python
model = models.Sequential()
model.add(Dense(12, activation = "relu", input_shape = (1,)))
model.add(Dense(8, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))
```

```python
model.compile(optimizer = "adam",
             loss = "binary_crossentropy",
             metrics = ["accuracy"])
```

```python
history1 = model.fit(x_train, y_train, epochs = 150,
                    batch_size = 128)
```

```python
model.summary()
```

```python
result = model.predict(x_test)
result
```

```python
score = model.evaluate(x_test, y_test, verbose = 0)
print("Test loss: ", score[0])
print("Test accuracy: ", score[1])
```

```python
plt.plot(history1.history["accuracy"])
plt.ylabel("Model Score")
plt.xlabel("Number of Epochs")
plt.show()
```

```python
churn["TotalCharges"] = pd.to_numeric(churn["TotalCharges"], errors = "coerce")
```

```python
X = pd.DataFrame(churn[["tenure","MonthlyCharges","TotalCharges"]])
Y = pd.DataFrame(churn["Churn"])
```

```python
Y.replace(("Yes", "No"), (1,0), inplace = True)
```

```python
X.info()
```

```python
print(X.shape)
print(Y.shape)
```

```python
x = X.apply(lambda x: ((x - x.min()) / (x.max() - x.min())))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test, x_train, y_test, y_train = train_test_split(X, Y,  test_size = .3)
```

```python
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
x_train = x_train.astype("float32")
x_test = x_test.astype("float32")
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
model = models.Sequential()
model.add(Dense(12, activation = "relu", input_shape = (3,)))
model.add(Dense(8, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))
```

```python
model.compile(optimizer = "adam",
             loss = "hinge",
             metrics = ["accuracy"])
```

```python
result3 = model.predict(x_test)
result3
```
