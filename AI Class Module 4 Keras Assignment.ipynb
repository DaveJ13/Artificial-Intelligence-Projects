```python
import tensorflow as tf
from keras import models
from keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
df.describe()
```

```python
df["Outcome"].value_counts()
```

```python
df_1 = df[["Pregnancies","Glucose","BloodPressure"]]
y = df["Outcome"]
```

```python
df_1.value_counts().isnull()
```

```python
y.value_counts().isnull()
```

```python
print(df_1.shape)
print(y.shape)
```

```python
x = df_1.apply(lambda x: ((x - x.min()) / (x.max() - x.min())))
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test, x_train, y_test, y_train = train_test_split(x, y,  test_size = .2)
```

```python
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
x_train = x_train.astype("float32")
x_test = x_test.astype("float32")
```

```python
output = 3
y_train = to_categorical(y_train, output)
y_test = to_categorical(y_test, output)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape
```

```python
model = models.Sequential()
model.add(Dense(8, activation = "relu", input_shape = (3,)))
model.add(Dense(3, activation = "sigmoid"))
```

```python
model.compile(optimizer = "sgd",
             loss = "categorical_crossentropy",
             metrics = ["accuracy"])
```

```python
model.fit(x_train, y_train,
         batch_size = 10,
         epochs = 100,
         verbose = 1,
         validation_data = (x_test, y_test))
```

```python
score = model.evaluate(x_test, y_test, verbose = 0)
print("Test loss: ", score[0])
print("Test accuracy: ", score[1])
```

```python
model.summary()
```
